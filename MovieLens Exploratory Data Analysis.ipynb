### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here
```

```python
rating_gr = rating.drop(columns=['user id', 'timestamp'])
rating_gr.head()
```

```python
rating_gr.info() #100,000rows
```

```python
rating_gr.nunique() #1682 unique movie ids
```

```python
#find the number of times each movie was rated
rating_gr2 =rating_gr.groupby(by=["movie id"]).count()
rating_gr2.head()
```

```python
rating_gr2.info() #1682 rows = number of unique movie ids
```

```python
#find the average rating per movie id
rating_gr3 =rating_gr.groupby(by=["movie id"]).mean()
rating_gr3.head()
```

```python
rating_gr2.rename(columns = {'rating':'cnt'}, inplace = True) 
rating_gr3.rename(columns = {'rating':'avg_rating'}, inplace = True) 
```

```python
#concat dataframes, rename columns, 
#filter to where total number of ratings is > 100, sort by average rating
rating_all = pd.concat([rating_gr2, rating_gr3], axis=1)
rating_all=rating_all[rating_all['cnt'] > 100]
rating_all = rating_all.sort_values(by=['avg_rating'],ascending=False)
rating_all=rating_all.head(25)
```

```python
#get the movie titles
movie_title = movie[['movie id', 'movie title']]
```

```python
rating_all2 = rating_all.reset_index()
final = rating_all2.merge(movie_title, how='inner', on='movie id')
final
```

*** Insights:  "A Close Shave" has 112 total ratings with the highest average rating of 4.5.

```python
user_movie_total = user_movie2['gender'].value_counts()
user_movie_total2 = user_movie_total.to_frame()
user_movie_total2 = user_movie_total2.rename(columns = {'gender':'cnt'})
user_movie_total2
```

```python
#Men watch more drama than women
user_movie21 = user_movie2.groupby('gender').sum()['Drama'].sort_values(ascending = False) 
user_movie21 = user_movie21.to_frame()
user_movie21
```

```python
drama = pd.concat([user_movie_total2,user_movie21], axis=1)
drama
```

```python
user_movie3 = user_movie2.groupby('gender').agg({'Drama':'sum'})
user_movie3
user_movie3.apply(lambda x: x/x.sum()*100).rename(columns = {'Drama':'% of total'})
```

```python
drama_pct = (drama.div(drama['cnt'], axis= 0) * 100)
drama_pct = drama_pct.rename(columns = {'Drama': '%total'})
drama_pct = drama_pct.drop(columns=['cnt'])
drama_pct 
```
